```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import acovf, acf, pacf,pacf_yw,pacf_ols
```

```python
df1 = pd.read_csv('Data/airline-passengers.csv', index_col='Month',parse_dates=True)
df1.index.freq = 'MS'
```

```python
df2 = pd.read_csv('Data/daily-total-female-births.csv', index_col='Date', parse_dates=True)
df2.index.freq='D'
```

```python
from pandas.plotting import lag_plot
```

```python
lag_plot(df1['Passengers'])
```

```python
lag_plot(df2['Births'])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(df1, lags=40);
```

```python
plot_acf(df2,lags=40);
```

```python
plot_pacf(df2,lags=40,title='Births');
```
